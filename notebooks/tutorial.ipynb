```python
from mjlab.scripts.train import launch_training, TrainConfig
from dataclasses import replace

config = TrainConfig.from_task("Mjlab-Tracking-Flat-Unitree-G1")

config = replace(
    config,
    registry_name="your-org/motions/motion-name",
    gpu_ids=[0],
    video=True,
    video_interval=1000,
)

launch_training("Mjlab-Tracking-Flat-Unitree-G1", config)
```
